```python
#Particular cleaning to movementNC_06-2020
pattern1 = r'(\d\.\d+)'
auxlist = []
dataL=[]
dataV=[]
for row in csvfiles['movementsNC_06-2020']['LANC.VALOR DESCRITIVO']:#This loop is to get the dates
    liste = re.findall(pattern1, str(row))
    auxlist.append(liste)

for n in range(0, len(auxlist)): #Loop to assign each date to each independent list
    dataL.append(auxlist[n][0])
    dataV.append(auxlist[n][1])

csvfiles['movementsNC_06-2020'].insert(0,'Data lanc', dataL, True) 
csvfiles['movementsNC_06-2020'].insert(1,'Data valor', dataV, True)

auxlist.clear()
for row in csvfiles['movementsNC_06-2020']['LANC.VALOR DESCRITIVO']:#Loop to get the information w/o the dates
    text = re.sub(pattern1,'',row)
    auxlist.append(text)
csvfiles['movementsNC_06-2020'].insert(2,'Descritivo',auxlist, True) #create a new colum with the new list, and delete the previous one
csvfiles['movementsNC_06-2020'].dtypes
csvfiles['movementsNC_06-2020'].drop(columns='LANC.VALOR DESCRITIVO', inplace=True)

pattern2=r' '
auxlist.clear()
for row in csvfiles['movementsNC_06-2020']['SALDO']:#This pretty much the same with the previous loops
    text = re.sub(pattern2,'',str(row))
    auxlist.append(text)
csvfiles['movementsNC_06-2020'].insert(5,'Saldo',auxlist,True)   
csvfiles['movementsNC_06-2020'].drop(columns='SALDO', inplace=True)
csvfiles['movementsNC_06-2020'].rename(columns={'DEBITO':'Debito','CREDITO': 'Credito'}, inplace=True)
```

```python
def remove_space(x):
    pattern =' '
    return re.sub(pattern,'',x)
```

```python
def date_format(x): #   M/D/Y
    pattern= '\.'
    return re.sub(pattern,'/',x) + "/2020"

def remove_date(x):
    pattern = ' \d\.\d+'
    return re.sub(pattern,'',x)
```

```python
def remove_numbers(x): #Apply only in Descritivo
    pattern='\d'
    return re.sub(pattern,'',x)
```

```python
#Functions to create column

def get_values(df): #1st this one, it will return you a nested list [[date][date]]
    datalist =[]
    pattern = ' \d\.\d+'
    data = re.findall(pattern, df)
    datalist.append(data[0])
    return datalist

def create_list(ListFromget_value): #2nd this one, it
    aulist =[]
    for block in ListFromget_value:
        for number in block:
            aulist.append(number)
    return aulist

def create_column(df,values): #3rd this one
    return df.insert(1,'Data valor', values,True)
```

```python
"""apply functions"""
csvfiles['movementsNC_06-2020']['Data lanc'] = csvfiles['movementsNC_06-2020']['Data lanc'].apply(date_format) 
csvfiles['movementsNC_06-2020']['Data valor'] = csvfiles['movementsNC_06-2020']['Data valor'].apply(date_format)
csvfiles['movementsNC_06-2020']['Descritivo'] = csvfiles['movementsNC_06-2020']['Descritivo'].apply(remove_numbers)
```
